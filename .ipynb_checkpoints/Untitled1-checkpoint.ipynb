```python
import cv2
import tensorflow as tf
import face_recognition
import numpy as np
import mediapipe as mp

path = r"Videos\oil-vid.mp4"
class_names = ['Acne', 'Blackheads', 'Burned', 'Clear_Skin', 'Dark_Circles', 'Dark_Spots', 'Droppy', 'Dry', 'Eye_Bags', 'Fine_Lines', 'Oily', 
               'Patches', 'Pigmentation', 'Pimples', 'Pores', 'Redness', 'Whiteheads', 'Wrinkles']
cap = cv2.VideoCapture(path)

# Loading the model
model = tf.keras.models.load_model("Model-Train.keras")

# Preprocess img function
IMG_SIZE = (224, 224)
def load_and_prep(image):
    img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    img = cv2.resize(img, IMG_SIZE)
    img = img / 255.0  # Normalize to [0, 1]
    return img

def get_face_image(frame, bbox):
    top, right, bottom, left = bbox
    face_image = frame[top:bottom, left:right]
    return face_image

# MediaPipe face mesh setup
mp_face_mesh = mp.solutions.face_mesh
face_mesh = mp_face_mesh.FaceMesh(static_image_mode=False, max_num_faces=1, min_detection_confidence=0.5)

while True:
    ret, img = cap.read()
    if not ret:
        cap = cv2.VideoCapture(path)
        ret, img = cap.read()

    img = cv2.resize(img, (800, 480), interpolation=cv2.INTER_AREA)

    # Detect faces using face_recognition
    face_locations = face_recognition.face_locations(img)

    if face_locations:
        top, right, bottom, left = face_locations[0]
        face_image = get_face_image(img, (top, right, bottom, left))
        img_pred = load_and_prep(face_image)
        pred_prob = model.predict(tf.expand_dims(img_pred, axis=0))[0]
        
        # Get class names and probabilities for the top predictions
        top_indices = pred_prob.argsort()[-3:][::-1]  # Get top 3 predictions
        filtered_indices = [idx for idx in top_indices if pred_prob[idx] > 0.005]  # Adjust the threshold as needed
        titles = [f"{class_names[i]}: {pred_prob[i]*100:.2f}%" for i in filtered_indices]
        title_text = ", ".join(titles)
        
        cv2.putText(img, text=title_text, org=(left, top - 10), fontFace=cv2.FONT_HERSHEY_COMPLEX, fontScale=0.6,
                    color=(0, 255, 0), thickness=2)
        cv2.rectangle(img, (left, top), (right, bottom), (0, 255, 0), 2)
        
        # Convert the BGR image to RGB and process with MediaPipe Face Mesh
        img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        results = face_mesh.process(img_rgb)

        if results.multi_face_landmarks:
            for face_landmarks in results.multi_face_landmarks:
                mp.solutions.drawing_utils.draw_landmarks(
                    image=img,
                    landmark_list=face_landmarks,
                    connections=mp_face_mesh.FACEMESH_TESSELATION,
                    landmark_drawing_spec=mp.solutions.drawing_utils.DrawingSpec(color=(0, 255, 0), thickness=1, circle_radius=1),
                    connection_drawing_spec=mp.solutions.drawing_utils.DrawingSpec(color=(0, 255, 0), thickness=1, circle_radius=1))
        
    cv2.imshow('Face Detection and Skin Condition Classification', img)

    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import time
import tensorflow as tf
import face_recognition

path = r"Videos\3182034-uhd_3840_2160_25fps.mp4"
class_names = ['Acne', 'Blackheads', 'Burned', 'Clear_Skin', 'Dark_Circles', 'Dark_Spots', 'Droppy', 'Dry', 'Eye_Bags', 'Fine_Lines', 'Oily', 
               'Patches', 'Pigmentation', 'Pimples', 'Pores', 'Redness', 'Whiteheads', 'Wrinkles']
cap = cv2.VideoCapture(path)
pTime = 0

# Loading the model
model = tf.keras.models.load_model("Model-Train.keras")

# Preprocess img function
IMG_SIZE = (224, 224)
def load_and_prep(image):
    img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    img = cv2.resize(img, IMG_SIZE)
    img = img / 255.0  # Normalize to [0, 1]
    return img

def get_face_image(frame, bbox):
    top, right, bottom, left = bbox
    face_image = frame[top:bottom, left:right]
    return face_image

while True:
    ret, img = cap.read()
    if not ret:
        cap = cv2.VideoCapture(path)
        ret, img = cap.read()

    img = cv2.resize(img, (800, 480), interpolation=cv2.INTER_AREA)

    # Detect faces using face_recognition
    face_locations = face_recognition.face_locations(img)

    if face_locations:
        top, right, bottom, left = face_locations[0]
        face_image = get_face_image(img, (top, right, bottom, left))
        img_pred = load_and_prep(face_image)
        pred_prob = model.predict(tf.expand_dims(img_pred, axis=0))
        pred_class = class_names[pred_prob.argmax()]
        title = f"Skin Type: {pred_class}, Prob: {pred_prob.max():.2f}%"
        cv2.putText(img, text=title, org=(left, bottom + 22), fontFace=cv2.FONT_HERSHEY_COMPLEX, fontScale=0.6,
                    color=(0, 255, 0), thickness=2)

    cTime = time.time()
    if (cTime - pTime) != 0:
        fps = 1 / (cTime - pTime)
        pTime = cTime

    cv2.putText(img, str(int(fps)), org=(15, 60), fontFace=cv2.FONT_HERSHEY_COMPLEX, fontScale=2,
                thickness=3, color=(255, 255, 255))
    cv2.imshow("Video", img)

    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```
